```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
df = pd.read_excel('crime rate.xlsx', names = ["Year", "Assault", "Homicide", "Robbery"])
```

```python
df.head()
```

```python
plt.style.use('seaborn-white')
```

```python
df.plot.line(x = 'Year', y = ['Assault', "Homicide", 'Robbery'], figsize = (8,6), title = "The crime rates have fallen gradually in Russia in the first two decades of the 21st century.", 
             colormap = "copper", grid = None, legend = True)
plt.suptitle("Russia to become a safer place in recent years", size = 14, fontweight = 'bold')
plt.legend(loc = "upper right")
plt.ylabel("cases per 100000 population")
plt.show();
```
